# Загрузка библиотек

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.preprocessing import LabelBinarizer

from sklearn.manifold import TSNE

from sklearn.cluster import AffinityPropagation, SpectralClustering, KMeans
```

# Определение функций

## Функция выделения признаков методом One-Hot 

```python
def encoder(df, label):
    jobs_encoder = LabelBinarizer()
    jobs_encoder.fit(df[label])
    transformed = jobs_encoder.transform(df[label])
    ohe_df = pd.DataFrame(transformed)
    
    return pd.concat([df, ohe_df], axis=1).drop([label], axis=1)
```

## Функция отображения граффика с метками

```python
def scat (X_embedded, SC):
    
    X_embedded_0 = X_embedded.loc[SC['name'] == 0]
    X_embedded_1 = X_embedded.loc[SC['name'] == 1]
    X_embedded_2 = X_embedded.loc[SC['name'] == 2]
    X_embedded_3 = X_embedded.loc[SC['name'] == 3]

    fig, ax = plt.subplots()

    ax.scatter(X_embedded_0[1], X_embedded_0[0],s = 500, marker='s')
    ax.scatter(X_embedded_1[1], X_embedded_1[0],s = 500, marker='X')
    ax.scatter(X_embedded_2[1], X_embedded_2[0],s = 500, marker='o')
    ax.scatter(X_embedded_3[1], X_embedded_3[0],s = 500, marker='p')

    fig.set_figwidth(12)
    fig.set_figheight(7)
    
    fig.patch.set_visible(False)
    #ax.axis('off')
    
    plt.legend(['Первый класс','Второй класс','Третий класс','Четвертый класс',],fontsize=12)

    plt.show()
```

## Вывод таблицы кластеров

```python
def tabl (y, SC):

    exp_1 = pd.concat([y, SC], axis=1)

    class_1 = pd.DataFrame(exp_1['название'].loc[exp_1['name'] == 0]).reset_index(drop=True)
    class_1.rename(columns={'название': 'class_1'}, inplace=True)

    class_2 = pd.DataFrame(exp_1['название'].loc[exp_1['name'] == 1]).reset_index(drop=True)
    class_2.rename(columns={'название': 'class_2'}, inplace=True)

    class_3 = pd.DataFrame(exp_1['название'].loc[exp_1['name'] == 2]).reset_index(drop=True)
    class_3.rename(columns={'название': 'class_3'}, inplace=True)

    class_4 = pd.DataFrame(exp_1['название'].loc[exp_1['name'] == 3]).reset_index(drop=True)
    class_4.rename(columns={'название': 'class_4'}, inplace=True)

    return pd.concat([class_1, class_2, class_3, class_4], axis=1)
```

## Описание признаков

название - название робота, данное ему производителем 

производитель - главная компания, организация, государственная структура принимавшая участие в разработке и создании робота

страна - страна в которой находится производитель и где производились основные разработки

год - год последнего официального изменения или модернизации робота

масса(кг) - масса робота выраженная в килограммах

скорость(км/ч) - максимальная скорость передвижения робота в пространстве

длина(м) - наибольная сторона робота из ширины и длины, выраженная в метрах

зона_действия - месьность или условия для функционирования в которых робот предназначен

животное - название живого организма, строение которого легло в основу робота

завершен - является проект закрытым или завершенным на момент составления таблицы

# Проверка на алгоритмах кластеризации

## Кластеризация методом распространения близости

```python
clustering = AffinityPropagation().fit(x)
clustering.labels_
```

```python
AP = pd.DataFrame(clustering.labels_)
AP.rename(columns={0: 'name'}, inplace=True)
```

## Спектральная кластеризация

```python
clustering = SpectralClustering(n_clusters=4, assign_labels="discretize", random_state=0).fit(x)
clustering.labels_
```

```python
SC = pd.DataFrame(clustering.labels_)
SC.rename(columns={0: 'name'}, inplace=True)
```

```python
scat(X_embedded, SC)
```

```python
tabl(y, SC)
```
